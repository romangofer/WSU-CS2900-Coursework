Part 2:

```python
import matplotlib.pyplot as plt
import json

extracted_temps = []
days = []

with open("forecast_for_ex4.json") as f:
    weekly_forecast = json.load(f)

properties = weekly_forecast["properties"]

temps = properties["periods"]

for t in temps:
         if t["number"] != 0:
             extracted_temps.append(t["temperature"])
             days.append(t["name"])
fig, ax = plt.subplots()
plt.style.use('seaborn')

ax.set_title("Forecast Temperatures Based on Time", fontsize = 18)

ax.set_xlabel("Time of the Week", fontsize = 15)
ax.set_ylabel('Temperature (F)', fontsize = 15)
ax.plot(days, extracted_temps, c='orange')
fig.autofmt_xdate()
plt.show()

# Sources Used:

# Demo Code from 6/9 Lecture


```
