```python
fav_nums = {'sam':56, 'ken':98, 'saul':6}
lookup = ('sam','ben','ken','paul','saul')

print('Favorite numbers:')

for key in lookup:
    if key not in fav_nums:
        input_value=int(input(f'Enter a favorite number for {key.title()}:'))
        fav_nums.update({key: input_value})

for key, value in fav_nums.items():
    print(f'\t{key.title():10}{value}')

fav_nums_listed = list(fav_nums.values())


added=sum(fav_nums_listed)
length=len(fav_nums_listed)
final=round(added/length)

print('Bonus - Average Favorite Number:')
print(f"\t{final}")

# Sources Used:

# In-class demos

# Stack Overflow - https://stackoverflow.com/questions/48234473/python-attributeerror-dict-object-has-no-attribute-append

# https://thispointer.com/python-4-ways-to-print-items-of-a-dictionary-line-by-line/

# Stack Overflow #2 - https://stackoverflow.com/questions/54573444/how-to-print-key-value-pairs-of-a-dict-as-an-aligned-table
```

```python
known_ips = ['12.45.23.1', '89.34.23.67', '67.45.23.43', '90.32.65.98']
def items(new_ips):
    for new_ip in new_ips:
        if new_ip not in known_ips and new_ip != '1.1.1.1':
           known_ips.append(new_ip) 

new_ips = ['45.23.98.12', '89.23.21.9', '67.45.23.43', '1.1.1.1']
items(new_ips)
new_ips = ['1.1.1.1']
items(new_ips)
new_ips = ['78.34.23.21', '90.43.65.1']
items(new_ips)

print('Known ips:')
for known_ip in known_ips:
   print(f"\t{known_ip}")

# Sources Used:

# In-class examples
```
